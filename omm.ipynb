```python
import numpy as np
import matplotlib.pyplot as plt
import math

#masa
m = 200
#dubina
h = 100
#koeficijent proporcionalnosti
c = 5
G = 9.81
#v0, teta

def foo(teta, v0):
    return (4*np.sin(teta)**3*np.cos(teta) - 3*h*c*np.sin(teta)**2*np.cos(teta)/(m*v0) - np.sin(teta)*np.cos(teta))


#x je ugao teta ogranicavamo ga na [0, pi/2]
# v cemo fiksirati pa posmatrati rezultate za razlicite v
#trazimo max funkcije
x = np.linspace(0, np.pi/2, 3000)
y = foo(x,30)
plt.figure(figsize=(15,10))
plt.plot(x,y, color='red')
max_y = np.max(y)
i = (y.tolist().index(max_y))
print(x[i]*180/np.pi)
print(y[i])
plt.plot(x[i], y[i], "ro")
plt.xlabel('x - ugao teta')
plt.ylabel('y - distanca')
plt.grid(axis = 'x')
plt.savefig('brzina_30.jpg')
plt.show()
```
